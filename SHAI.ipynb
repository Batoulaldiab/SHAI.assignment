**4**\- Create histograms or bar charts to visualize the distribution of salaries, and use pie charts to represent the proportion of employees in different departments.


```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('Salaries.csv')
names = df['EmployeeName'][:10]
salaries = df['TotalPay'][:10]
plt.plot(names, salaries, marker='o')
plt.title('Salary distribution')
plt.xlabel('Salaries')
plt.ylabel('Names')
plt.show()
# قمت بتثميل بيانات اول عشر فقط لان عدد الموظفين كبير جدا
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('Salaries.csv')
department_percentages = data['JobTitle'].value_counts(normalize=True).head(10) * 100
plt.figure(figsize=(8, 8))
plt.pie (department_percentages, labels=department_percentages.index, autopct='%1.1f%%', startangle=140)
plt.title("Number of employees in each department (Top 10)")
plt.show()
# قمت بتثميل بيانات أعلى عشر فقط لان عدد الاقسام كبير جداا
```

**5**\- Group the data by one or more columns and calculate summary statistics for each group, and compare the average salaries across different groups.


```python
import pandas as pd
data = pd.read_csv('Salaries.csv')
average_salary_by_job = data.groupby('JobTitle')['TotalPay'].mean()
print("Average salaries per job:")
print(average_salary_by_job)

average_salary_by_JobTitle = data.groupby('JobTitle')['TotalPay'].mean()

print("Average salaries for each JobTitle:")
print(average_salary_by_JobTitle)
```

**6** \- Identify any correlation between salary and another numerical column, and plot a scatter plot to visualize the relations


```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('Salaries.csv')

x_column = 'TotalPay'
y_column = 'OtherPay'

plt.figure(figsize=(10, 6))
plt.scatter(data[x_column], data[y_column], alpha=0.5)
plt.title(f"Scatter Plot: {x_column} vs {y_column}")
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.show()
```
